# 1. ¿Cuantas posibilidades hay sin tener en cuenta las restricciones?

No tener en cuenta las restricciones significa que los números y las operaciones se pueden repetir. En este caso las variaciones de los operaciones serían:

$V_{operaciones} = m^{n} = 4^4 = 256$

En el caso de los números, sería:

$V_{numeros} = m^n = 9^5 = 59049$

Finalmente, tenemos que en total:

$V_{total} = V_{operaciones} * V_{numeros} = 256 * 59049 = 15116544$

## 1.1 ¿Cuantas posibilidades hay teniendo en cuenta todas las restricciones?

Para poder calcular la cantidad de posibilidades teniendo en cuenta las restricciones se debe realizar la combinatoria.

La cantidad de números que se pueden tomar son 5 de un conjunto de 9 valores. Es decir:

$\frac{n!}{k!(n - k)!}=\frac{9!}{5!(9 - 5)!}=\frac{9!}{5!4!}=\frac{9*8*7*6*5!}{5!*4*3*2}=\frac{9*8*7*6}{4*3*2}=126$

Ahora calculamos las combinaciones de los 5 valores:

$n! = 5! = 120$

Finalmente, calculamos el total:

$126 * 120 = 15120$

Con respecto a los signos, tenemos que encontrar la combinatoria sin repetición de los signos:

$n!=4!=24$

Y para calcular el total de iteraciones, realizamos la multiplicación:

$T_{iteraciones} = (120 * 126) * 24 = 362880$ iteraciones

# 5. (*) Diseña un algoritmo que mejore la complejidad del algortimo por fuerza bruta. Argumenta porque crees que mejora el algoritmo por fuerza bruta



```
from itertools import permutations, combinations
import time


# cálculo de las permitaciones
nums = ('1', '2', '3', '4', '5', '6', '7', '8', '9')
operations = ('+', '-', '*', '/')
_final = [x for x in permutations(nums, 5)]
_operations = [x for x in permutations(operations, 4)]

# algoritmo para evaluar las expresiones y encontrar las operaciones que cumplen
# con expresión
def find_sum(to_find):
  solutions = []

  for data in _final:
    for operator in _operations:
        combination = f'{data[0]}{operator[0]}{data[1]}{operator[1]}{data[2]}{operator[2]}{data[3]}{operator[3]}{data[4]}'
        result = eval(combination)
  
        if result % 1 == 0:
          if to_find == result:
            solutions.append(combination)

  return solutions

print(find_sum(4))
```

### 5.1 (*) Calcula la complejidad del algoritmo

La primera parte del algoritmo se usa para calcular las permutaciones de los datos de entrada, este algoritmo tiene complejidad $O(n!)$ en tiempo y $O(n)$ en espacio.

Por otro lado, en la segunda parte del ejercicio se realizan las iteraciones para combinar las permutaciones y evaluar la expresión. Esta parte tiene complejidad $O(n^2)$ en tiempo y $O(n)$ en espacio.

Al analizar las complejidades se puede ver como la complejidad del algoritmo completo tiene a $O(n!)$ en tiempo y $O(n)$ en espacio.

## 6 Según el problema (y tenga sentido), diseña un juego de datos de entrada aleatorios

```
import random

nums = ['1', '2', '3', '4', '5', '6', '7', '8', '9']
operations = ['+', '-', '*', '/']

def get_random_values(total_number_combination, total_operations):
  random_nums = [random.sample(nums, 5) for _ in range(0, total_number_combination)]
  random_operations = [random.sample(operations, 4) for _ in range(0, total_operations)]

  return (random_nums, random_operations)
```
